Working Algorithm

```python
reader = pd.read_csv('Nifty_50_companies_list.csv')
investment = 100000
profitpr = []


investment_standard_deviation = []
min_investment = []
# print(reader['Company'])

# file = open('Nifty_50_companies_list.csv', 'r')
# reader = csv.reader(file)
for row in reader['Symbol']:
    # print(row[2])
    cp = str(row) + ".NS"
    print(str(cp))
    data = yf.download(cp, Starttime, Endtime, interval = Interval)
    ans = 0
    neg = 0
    san = 0
    flag = 0
    investment_list = []
    print("step 1 done")
    for i in data.index:
        # if(str(i.time()) == "09:30:00"):
            # print("first high " + str(data.loc[i]['High']))
            # print("second close " + str(data.loc[i + timedelta(minutes=Interval)]['Close']))
            # print("third open " + str(data.loc[i+timedelta(minutes=Interval*2)]['Open']))
            # print("third close " + str(data.loc[i + timedelta(minutes=Interval*2)]['Close']))
        if flag == 0:
            print("step 2 ongoing")
            flag = 1
            # print(data.loc[i+timedelta(minutes=Interval*2)]['High'])
        if(str(i.time()) == "09:15:00" and data.loc[i]['High'] < data.loc[i + timedelta(minutes=Interval)]['Close'] and data.loc[i+timedelta(minutes=Interval*2)]['Open'] < data.loc[i + timedelta(minutes=Interval*2)]['Close']):
            
            # print(((data.loc[i + timedelta(minutes=Interval*2)]['Close'] - data.loc[i+timedelta(minutes=Interval*2)]['Open'])/data.loc[i+timedelta(minutes=Interval*2)]['Open'])*100)
            quantity = 60/(data.loc[i+timedelta(minutes=Interval*2)]['Open']*((data.loc[i + timedelta(minutes=Interval*2)]['Close'] - data.loc[i+timedelta(minutes=Interval*2)]['Open'])/data.loc[i+timedelta(minutes=Interval*2)]['Open']))
            tmp_profit = abs((data.loc[i + timedelta(minutes=Interval*2)]['Close'] - data.loc[i+timedelta(minutes=Interval*2)]['Open'])/data.loc[i+timedelta(minutes=Interval*2)]['Open'])
            investment1 = quantity * data.loc[i+timedelta(minutes=Interval*2)]['Open']
            if investment1 <= investment:
                # print("investment = " + str(investment1))
                ans = ans + 1
                investment_list.append(investment1)
        elif(str(i.time()) == "09:15:00" and data.loc[i]['Low'] > data.loc[i + timedelta(minutes=Interval)]['Close'] and data.loc[i+timedelta(minutes=Interval*2)]['Open'] > data.loc[i + timedelta(minutes=Interval*2)]['Close']):
            
            quantity = 60/(data.loc[i+timedelta(minutes=Interval*2)]['Open']*((data.loc[i + timedelta(minutes=Interval*2)]['Close'] - data.loc[i+timedelta(minutes=Interval*2)]['Open'])/data.loc[i+timedelta(minutes=Interval*2)]['Open']))
            tmp_profit = abs((data.loc[i + timedelta(minutes=Interval*2)]['Close'] - data.loc[i+timedelta(minutes=Interval*2)]['Open'])/data.loc[i+timedelta(minutes=Interval*2)]['Open'])
            investment1 = quantity * data.loc[i+timedelta(minutes=Interval*2)]['Open']
            if investment1 <= investment:
                # print("investment = " + str(investment1))
                san = san + 1
                investment_list.append(abs(investment1))

            
            # print(((data.loc[i + timedelta(minutes=Interval*2)]['Close'] - data.loc[i+timedelta(minutes=Interval*2)]['Open'])/data.loc[i+timedelta(minutes=Interval*2)]['Open'])*100)
        elif(str(i.time()) == "09:15:00"):
        
            neg = neg + 1
        #print("step 2 half over")
    if len(investment_list)>0:
        temp = min(investment_list)
    else:
        temp = 0
    # print(temp)
    min_investment.append(temp)
    # print(min_investment)
    profitpr.append(tmp_profit)

    if(len(investment_list) > 1):
        investment_standard_deviation.append(abs(statistics.stdev(investment_list)))
        # min_investment.append(temp)
    else:
        investment_standard_deviation.append(-1)
        # min_investment.append(temp)
    # print(investment_standard_deviation)
    print(min_investment)
    # print(tmp)
    print(ans)
    print(san)
    print(neg)

print("step 2 over")

sheet['standard_deviation'] = investment_standard_deviation
sheet['min_investment'] = min_investment
sheet['profit_pr'] = profitpr
```

```python
sheet = sheet.sort_values('standard_deviation')
sheet.reset_index(drop=True, inplace=True)
sheet = sheet.loc[0:26]
sheet = sheet.sort_values(by='profit_pr', ascending=False)
sheet.reset_index(drop=True, inplace=True)
print(sheet)
    
```
